```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
import requests

# Step 1: Fetching Historical BTC/USDT Data from CoinGecko
def fetch_historical_data(days=30):
    url = f"https://api.coingecko.com/api/v3/coins/bitcoin/market_chart?vs_currency=usd&days={days}"
    try:
        response = requests.get(url)
        data = response.json()
        prices = data.get("prices", [])
        if prices:
            df = pd.DataFrame(prices, columns=["timestamp", "price"])
            df['timestamp'] = pd.to_datetime(df['timestamp'], unit='ms')
            df.set_index('timestamp', inplace=True)
            return df
        else:
            print("Failed to fetch prices from CoinGecko.")
            return None
    except Exception as e:
        print(f"Error fetching data: {e}")
        return None

# Step 2: Dynamically Adjust Parameters
def estimate_parameters(data):
    data['Log_Returns'] = np.log(data['price'] / data['price'].shift(1)).dropna()
    X_t_minus_1 = data['price'].shift(1).dropna().values.flatten()
    X_t = data['price'][1:].values.flatten()

    try:
        regression_result = linregress(X_t_minus_1, X_t)
        theta = 1 - regression_result.slope
        mu = regression_result.intercept / (1 - regression_result.slope)
        sigma = data['Log_Returns'].std()
        return theta, mu, sigma
    except Exception as e:
        print(f"Error estimating parameters: {e}")
        return None, None, None

# Step 3: Calculate Expected Exit Time with Improved Handling
def calculate_exit_time(theta, mu, sigma, val, vah):
    try:
        term1 = (1 / theta) * np.log((vah - mu) / (mu - val))
        term2 = (sigma ** 2) / (2 * theta ** 3) * ((1 / (vah - mu) ** 2) + (1 / (mu - val) ** 2))
        exit_time = term1 + term2

        if np.isnan(exit_time) or exit_time < 0:
            return "More than 1 hour"
        elif exit_time < 0.5:
            return "Less than 30 minutes"
        elif exit_time < 1:
            return "Within 1 hour"
        else:
            return "More than 1 hour"
    except Exception as e:
        print(f"Error calculating exit time: {e}")
        return "Unknown"

# Step 4: Dynamic Stop-Loss and Take-Profit with ROI Calculation
def backtest_ou_strategy(data):
    theta, mu, sigma = estimate_parameters(data)
    if theta is None:
        print("Error in parameter estimation.")
        return

    initial_capital = 1000
    capital = initial_capital
    position = None
    take_profit_threshold = 0.003  # 0.3% target for take-profit
    stop_loss_threshold = 0.002    # 0.2% target for stop-loss

    data['Position'] = 0
    data['Capital'] = capital
    data['Predicted'] = np.nan

    for i in range(1, len(data)):
        price = data['price'].iloc[i]
        exit_time = calculate_exit_time(theta, mu, sigma, val=mu - 30, vah=mu + 30)

        if position is None:
            if exit_time == "Less than 30 minutes" or exit_time == "Within 1 hour":
                # Enter long position
                entry_price = price
                position = "long"
                print(f"Buying at {entry_price}")
                data['Predicted'].iloc[i] = entry_price
            continue

        # Dynamic stop-loss and take-profit
        if position == "long":
            if price >= entry_price * (1 + take_profit_threshold):
                capital += (price - entry_price)
                position = None
                print(f"Take Profit: Closing Long at {price}, Capital: {capital}")
                data['Predicted'].iloc[i] = price
            elif price <= entry_price * (1 - stop_loss_threshold):
                capital -= (entry_price - price)
                position = None
                print(f"Stop Loss: Closing Long at {price}, Capital: {capital}")
                data['Predicted'].iloc[i] = price

    final_roi = ((capital - initial_capital) / initial_capital) * 100
    print(f"Final Capital: {capital}")
    print(f"Total Return: {final_roi}%")
    return capital, final_roi

# Step 5: Enhanced Visualization
def plot_backtest(data):
    plt.figure(figsize=(12, 6))
    plt.plot(data.index, data['price'], label="Actual BTC Price", color='blue', alpha=0.8)
    plt.scatter(data.index, data['Predicted'], color='red', marker='o', label='Predicted Price Movements')
    plt.axhline(y=data['price'].mean() + 30, color='orange', linestyle='--', label='VAH')
    plt.axhline(y=data['price'].mean() - 30, color='green', linestyle='--', label='VAL')
    plt.axhline(y=data['price'].mean(), color='purple', linestyle='-', label='POC (Mu)')
    plt.xlabel("Time")
    plt.ylabel("Price (USD)")
    plt.title("BTC/USDT Price Movements: Actual vs. Predicted with Value Areas")
    plt.legend()
    plt.grid(True)
    plt.show()

# Step 6: Running the Enhanced Backtest
historical_data = fetch_historical_data(days=30)
if historical_data is not None:
    final_capital, roi = backtest_ou_strategy(historical_data)
    plot_backtest(historical_data)
    print(f"Final ROI: {roi}%")

# Step 7:
import numpy as np

# Function to estimate expected exit time using the OU model with adjustments
def calculate_expected_exit_time(theta, mu, sigma, val, vah):
    a = val  # Support (VAL)
    b = vah  # Resistance (VAH)

    try:
        # Ensure input values for log and division are valid
        buffer = 1e-5  # Small buffer to prevent log issues
        if (b - mu) <= buffer or (mu - a) <= buffer:
            print("Invalid input for exit time calculation; returning NaN")
            return np.nan

        term1 = (1 / theta) * np.log(max((b - mu), buffer) / max((mu - a), buffer))
        term2 = (sigma ** 2) / (2 * theta ** 3) * ((1 / (max((b - mu), buffer) ** 2)) + (1 / (max((mu - a), buffer) ** 2)))

        # Expected exit time in hours
        expected_exit_time = term1 + term2

        # Convert to minutes for shorter-term trading relevance
        return expected_exit_time * 60  # Convert hours to minutes

    except Exception as e:
        print(f"Error calculating expected exit time: {e}")
        return np.nan

# Example parameters (ensure these are correctly set before using the function)
theta = 0.5         # Speed of mean reversion (example value)
mu = 67000          # Long-term mean (POC)
sigma = 0.002       # Volatility
val = 66000         # Support level (VAL)
vah = 68000         # Resistance level (VAH)

# Calculate expected exit time
exit_time_minutes = calculate_expected_exit_time(theta, mu, sigma, val, vah)
print(f"Expected time spent between VAL and VAH: {exit_time_minutes:.2f} minutes")

# Step 8:
```
